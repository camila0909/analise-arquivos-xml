# Aprendendo a extrair informações de arquivos .xml

```python
import xml.etree.ElementTree as et 
from xml.etree.ElementTree import Element

arquivo = et.parse("versao_teste_marcas.xml") # Lê o arquivo
revista = arquivo.getroot() # Chama o elemento raíz
```

## 1. Conhecendo a tag do elemento raíz e seus atributos

```python
revista.tag # Tag raíz
```

```python
revista.attrib # Atributos da Tag
```

```python
len(revista) # Quantidade de filhos diretos que essa tag possui
```

### 2.1 Criando uma lista de atributos (dicionários)

```python
processos = []

for filho in revista:
    processo = filho.attrib
    processos.append(processo)
    
print(processos)
```

### 2.2 Criando uma lista de nºs de processos

```python
processos = []

for filho in revista.findall('processo'):
    processo = filho.get('numero')
    processos.append(processo)
    
print(processos)
```

#### 3.1.1 Despachos

```python
for filho in revista: # Processos
    for filhos in filho.findall('despachos'): # Despachos
        for neto_despachos in filhos:
            neto_despachos = neto_despachos

print(neto_despachos.tag, neto_despachos.attrib) # Para retornar todos os despachos, basta incluir o print dentro de for.
```

```python
despachos = []

for filho in revista.findall('processo'):
    for filhos in filho.findall('despachos'):
        for neto_despachos in filhos.findall('despacho'):
            despacho = neto_despachos.attrib['nome']
            despachos.append(despacho)

despachos # Gera uma lista com todos os 'nome' em despachos do documentos
```

#### 3.1.2 Titulares

```python
for filho in revista: # Processos
    for filhos in filho.findall('titulares'): # Titulares
        for neto_titulares in filhos:
            neto_titulares = neto_titulares

print(neto_titulares.tag, neto_titulares.attrib)
```

```python
Titulares = []

for filho in revista.findall('processo'):
    for filhos in filho.findall('titulares'):
        for neto_titulares in filhos.findall('titular'):
            titular = neto_titulares.attrib['nome-razao-social']
            Titulares.append(titular)
            
Titulares # Gera uma lista com todos os 'nome-razao-social' em Titulares do documentos
```

#### 3.1.3 Marcas

```python
for filho in revista: # Processos
    for filhos in filho.findall('marca'): # Marca
        for neto_marca in filhos:
            neto_marca = neto_marca

print(neto_marca.tag, neto_marca.attrib)
```

```python
marcas = []

for processo in revista.findall('processo'): # Processos
    for marca in processo.findall('marca'):
        for nome in marca.findall('nome'):
            nome = nome.text
            marcas.append(nome)   
    
marcas #Deveria retornar uma lista com o nome das marcas, mas não consegui identificar o erro. =/
```

#### 3.1.4 Classes-Vienna

```python
for filho in revista: # Processos
    for filhos in filho.findall('classes-vienna'): # classes-vienna
        for neto_classes_vienna in filhos:
            neto_classes_vienna = neto_classes_vienna

print(neto_classes_vienna.tag, neto_classes_vienna.attrib)
```

#### 3.1.5 Lista-Classe-Nice

```python
for filho in revista: # Processos
    for filhos in filho.findall('lista-classe-nice'): # lista-classe-nice
        for neto_lista_classe_nice in filhos:
            neto_lista_classe_nice = neto_lista_classe_nice

print(neto_lista_classe_nice.tag, neto_lista_classe_nice.attrib)
```
